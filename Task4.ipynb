#Problem 1: Mobile Recharge Validation System

```python
def validate_recharge(amount):
    valid_plans = [199, 299, 399, 599]

    if amount < 50:
        print("Recharge amount must be at least ₹50.")
        return False
    elif amount not in valid_plans:
        print("Invalid recharge plan selected.")
        return False
    else:
        print("Recharge successful!")
        return True


# Retry system using while loop
while True:
    try:
        user_amount = int(input("Enter recharge amount: ₹"))

        if validate_recharge(user_amount):
            break   # Exit loop if recharge is valid

        print("Please try again.\n")

    except ValueError:
        print("Invalid input. Please enter a numeric value.\n")
```

#Problem 2: Inventory Reorder Alert System

```python
def check_inventory(products):
    for product, stock in products.items():
        if stock < 15:
            print(f"{product}: Reorder Alert (Stock = {stock})")
        else:
            print(f"{product}: Stock OK (Stock = {stock})")


# Sample dictionary
inventory = {
    "Rice": 10,
    "Sugar": 25,
    "Oil": 5,
    "Wheat": 18,
    "Salt": 12
}

# Call function
check_inventory(inventory)
```

#Problem 3: Student Result Processing

```python
def process_result(marks):
    total = 0

    # Calculate total using loop
    for mark in marks:
        total += mark

    average = total / len(marks)

    print("Average Marks:", average)

    if average >= 50:
        print("Result: Pass")
        return "Pass"
    else:
        print("Result: Fail")
        return "Fail"


# Example usage
student_marks = [60, 70, 55, 40, 65]
result = process_result(student_marks)
```

#Problem 4: Cab Fare Estimator with Retry Option


```python
def calculate_fare(distance, peak):
    base_fare = 50
    per_km_rate = 12

    total_fare = base_fare + (per_km_rate * distance)

    if peak.lower() == "yes":
        total_fare += total_fare * 0.25   # Add 25% extra

    return total_fare


# Retry system using while loop
while True:
    try:
        distance = float(input("Enter distance in km: "))

        if distance <= 0:
            print("Distance must be greater than 0.\n")
            continue

        peak_hour = input("Is it peak hour? (yes/no): ")

        if peak_hour.lower() not in ["yes", "no"]:
            print("Invalid input for peak hour.\n")
            continue

        fare = calculate_fare(distance, peak_hour)
        print(f"Total Fare: ₹{fare}")
        break   # Exit loop if successful

    except ValueError:
        print("Invalid input. Please enter numeric value for distance.\n")
```

#Problem 5: Employee Attendance Eligibility Checker


```python
def check_eligibility(attendance):
    present_count = 0

    # Count present days using loop
    for day in attendance:
        if day == "P":
            present_count += 1

    total_days = len(attendance)
    attendance_percentage = (present_count / total_days) * 100

    print("Attendance Percentage:", attendance_percentage)

    if attendance_percentage >= 75:
        return "Eligible"
    else:
        return "Not Eligible"


# Example usage
attendance_record = ["P", "P", "A", "P", "A", "P", "P"]
result = check_eligibility(attendance_record)
print("Result:", result)
```

#Problem 6: Password Strength Checker


```python
def check_password_strength(password):
    if len(password) < 8:
        return "Weak Password (Minimum length is 8)"

    has_digit = False
    has_special = False

    # Loop over string
    for char in password:
        if char.isdigit():
            has_digit = True
        if char in "@#$":
            has_special = True

    # Check conditions
    if has_digit and has_special:
        return "Strong Password"
    else:
        return "Weak Password (Must contain at least one digit and one special character)"


# Example usage
user_password = input("Enter your password: ")
result = check_password_strength(user_password)
print(result)
```
