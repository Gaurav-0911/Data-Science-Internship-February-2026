#Problem 1: Social Media – Post Engagement Analyzer

```python
def analyze_post_engagement(likes_list):
    total_likes = 0

    # Loop to calculate total likes
    for likes in likes_list:
        total_likes += likes

    print("Total Likes:", total_likes)

    # Condition to check post status
    if total_likes >= 1000:
        print("Post Status: Viral Post")
    else:
        print("Post Status: Normal Engagement")


# Example Data
likes_data = [200, 300, 150, 400]

analyze_post_engagement(likes_data)
```

#Problem 2: Healthcare – Medicine Stock Alert System

```python
def check_medicine_stock(stock):

    print("Medicine Stock:", stock)

    # Conditional Logic
    if stock < 10:
        print("Status: Low Stock Alert")
    else:
        print("Status: Stock Sufficient")


# Example Data
medicine_stock = 6

check_medicine_stock(medicine_stock)
```

#Problem 3: Agriculture – Rainfall Adequacy Checker

```python
def check_rainfall(rainfall_data, required_level):

    total_rainfall = 0

    # Loop to calculate total rainfall
    for rain in rainfall_data:
        total_rainfall += rain

    # Calculate average rainfall
    average_rainfall = total_rainfall / len(rainfall_data)

    print("Average Rainfall:", int(average_rainfall))

    # Conditional check
    if average_rainfall >= required_level:
        print("Rainfall Status: Adequate Rainfall")
    else:
        print("Rainfall Status: Inadequate Rainfall")


# Example Data (Seasonal Rainfall in mm)
seasonal_rainfall = [60, 75, 80, 73]
required_rainfall_level = 70

check_rainfall(seasonal_rainfall, required_rainfall_level)
```

#Problem 4: Social Media – Duplicate Account Detection

```python
def detect_duplicate_accounts(usernames):

    unique_usernames = set(usernames)

    # Compare lengths
    if len(unique_usernames) < len(usernames):
        print("Duplicate Accounts Found: Yes")
    else:
        print("Duplicate Accounts Found: No")


# Example Data
accounts = ["user1", "user2", "admin", "user1", "guest"]

detect_duplicate_accounts(accounts)
```

#Problem 6: Agriculture – Premium Crop Price Filter

```python
def filter_premium_crops(prices):

    premium_crops = []

    # Loop for filtering
    for price in prices:
        if price > 2000:
            premium_crops.append(price)

    print("Premium Crops:", premium_crops)


# Example Data (Crop Prices in ₹)
crop_prices = [1500, 2500, 1800, 3200, 1900]

filter_premium_crops(crop_prices)
```

#Problem 7: System Monitoring – Application Health Checker

```python
def check_application_health(error_count):

    print("Error Count:", error_count)

    # Conditional Statements
    if error_count == 0:
        print("System Status: Healthy")
    elif error_count <= 5:
        print("System Status: Minor Issues")
    else:
        print("System Status: Critical Issues")


# Example Data
errors = 7

check_application_health(errors)
```

#Problem 8: Banking – Daily Transaction Limit Checker

```python
def check_transaction_limit(amount):

    daily_limit = 50000

    print("Transaction Amount:", amount)

    # Conditional Logic
    if amount <= daily_limit:
        print("Transaction Status: Approved")
    else:
        print("Transaction Status: Rejected")


# Example Data
transaction_amount = 60000

check_transaction_limit(transaction_amount)
```

#Problem 9: E-Learning – Student Attendance Eligibility System


```python
def check_exam_eligibility(attendance_list):

    total_classes = len(attendance_list)
    attended_classes = 0

    # Loop to count attended classes
    for status in attendance_list:
        if status == 1:   # 1 means Present
            attended_classes += 1

    # Calculate percentage
    attendance_percentage = (attended_classes / total_classes) * 100

    print("Attendance Percentage:", attendance_percentage)

    # Eligibility Check
    if attendance_percentage >= 75:
        print("Exam Eligibility: Eligible")
    else:
        print("Exam Eligibility: Not Eligible")


# Example Data (1 = Present, 0 = Absent)
attendance_data = [1, 1, 1, 0, 1]   # 4 out of 5 classes attended

check_exam_eligibility(attendance_data)
```

#Problem 10: Smart Electricity Bill Analyzer


```python
def calculate_electricity_bill(units):

    bill = 0

    # Billing Slabs
    if units <= 100:
        bill = units * 3

    elif units <= 200:
        bill = (100 * 3) + ((units - 100) * 5)

    else:
        bill = (100 * 3) + (100 * 5) + ((units - 200) * 7)


    # Usage Classification
    if bill < 500:
        status = "Low Usage"
    elif bill <= 1500:
        status = "Moderate Usage"
    else:
        status = "High Usage"

    return bill, status


# Example Data
units_consumed = 250

total_bill, usage_status = calculate_electricity_bill(units_consumed)

print("Units Consumed:", units_consumed)
print("Total Bill: ₹", total_bill)
print("Usage Level:", usage_status)
```
