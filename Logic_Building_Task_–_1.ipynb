#1. User Login Check

#2. Pass / Fail Analyzer

```python
#list of marks that is given
marks=[45,78,90,33,60]
pass_count=0
fail_count=0

#each student's mark checkup
for mark in marks:
  if mark >= 50:
    pass_count += 1
  else:
    fail_count += 1

#Getting the final result output
print("Total Students : ", len(marks))
print("Total Passed : ", pass_count)
print("Total Failed : ", fail_count)
```

#3. Simple Data Cleaner

```python
#Given Unstructured/Un-cleaned List
names=[" Alice " ,"bob" , "CHARLIE"]

#Declaration
cleaned_names=[]

#Cleaning Process
for name in names:
  cleaned_name=name.strip().lower() #extra space removal & lowercasing
  cleaned_names.append(cleaned_name) #to cleaned tect append one by one
  print("\nCleaned Name: ", cleaned_names) # finally cleaned list formation
```

#4. Message Length Analyzer

```python
#list of Given messages
messages = ["Hi", "Welcome to the platform", "OK"]

for message in messages:
  length=len(message)
  print("Message :",message)
  print("length: ",length)
  if length > 10:
    print("Longer than 10 characters") #required flag message
  else:
    print("Shorter message")
  print("=======================================================")#used as seperator
```
