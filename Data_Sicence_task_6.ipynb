#Problem Statement 1:Smart Parking Lot Management System

```python
def smart_parking_system(capacity, logs):
    current_parked = 0
    peak_usage = 0

    for log in logs:
        if log == "IN":
            current_parked += 1
        elif log == "OUT" and current_parked > 0:
            current_parked -= 1

        if current_parked > peak_usage:
            peak_usage = current_parked

    print("Currently Parked Vehicles:", current_parked)

    if current_parked > capacity:
        print("Parking Status: Full (Alert! Capacity Exceeded)")
    else:
        print("Parking Status: Available")

    print("Peak Parking Usage:", peak_usage)


capacity = 50
logs = ["IN", "IN", "IN", "OUT", "IN", "IN", "OUT"]

smart_parking_system(capacity, logs)
```

# Problem Statement 2:Online Food Delivery Time Estimator

```python
def estimate_delivery_time(distance, traffic, weather):
    # Base time: 5 minutes per km
    eta = distance * 5

    # Traffic adjustment
    if traffic.lower() == "high":
        eta += 15
    elif traffic.lower() == "medium":
        eta += 10
    elif traffic.lower() == "low":
        eta += 5

    # Weather adjustment
    if weather.lower() == "rainy":
        eta += 10
    elif weather.lower() == "stormy":
        eta += 20
    elif weather.lower() == "clear":
        eta += 0

    print("Estimated Delivery Time:", eta, "minutes")


# Sample Input
distance = 6
traffic = "High"
weather = "Rainy"

estimate_delivery_time(distance, traffic, weather)
```

# Problem Statement 3:Movie Theatre Seat Occupancy Analyzer


```python
def movie_seat_analyzer(total_seats, booked_seats):
    booked_count = len(booked_seats)

    # Calculate occupancy percentage
    occupancy = (booked_count / total_seats) * 100

    print("Occupancy:", int(occupancy), "%")

    # Determine show status
    if occupancy == 100:
        print("Show Status: Housefull")
    elif occupancy >= 75:
        print("Show Status: Almost Full")
        print("Suggestion: Consider Opening Additional Show")
    else:
        print("Show Status: Seats Available")


# Sample Input
total_seats = 200
booked_seats = [1] * 150   # 150 booked seats

movie_seat_analyzer(total_seats, booked_seats)
```

# Problem Statement 4:Cloud Server Load Classification System


```python
def classify_server_load(cpu_readings):
    # Calculate average CPU usage
    total = 0
    for usage in cpu_readings:
        total += usage

    average = total / len(cpu_readings)

    print("Average CPU Load:", int(average), "%")

    # Classify server load
    if average < 50:
        print("Server Status: Normal")
    elif 50 <= average <= 80:
        print("Server Status: Warning")
    else:
        print("Server Status: Critical")


# Sample Input
cpu_readings = [45, 60, 70, 85, 90]

classify_server_load(cpu_readings)
```

#Problem Statement 5:Smart Classroom Resource Usage Monitor


```python
def smart_classroom_monitor(resource_usage):
    overused = []

    # Check for overuse (threshold: more than 8 hours)
    for resource, hours in resource_usage.items():
        if hours > 8:
            overused.append(resource)

    # Display results
    if overused:
        print("Overused Resources:", ", ".join(overused))
        print("Energy Alert: Yes")
    else:
        print("Overused Resources: None")
        print("Energy Alert: No")


# Sample Input
resource_usage = {
    "Projector": 6,
    "AC": 9,
    "Lights": 4
}

smart_classroom_monitor(resource_usage)
```

#Problem Statement 6: Online Event Registration Capacity Controller

```python
def event_registration_controller(capacity, registrations):
    confirmed = 0
    waitlisted = 0

    # Track registrations and prevent overbooking
    for i in range(registrations):
        if confirmed < capacity:
            confirmed += 1
        else:
            waitlisted += 1

    print("Confirmed Registrations:", confirmed)
    print("Waitlisted Users:", waitlisted)

    if confirmed >= capacity:
        print("Registration Status: Closed")
    else:
        print("Registration Status: Open")


# Sample Input
capacity = 100
registrations = 105

event_registration_controller(capacity, registrations)
```
